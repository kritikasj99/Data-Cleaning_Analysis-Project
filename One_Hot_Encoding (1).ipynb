**FILLING THE GENDER MISSING VALUES BY THE MODE**

```python
dataset["Gender"].fillna(dataset["Gender"].mode()[0],inplace=True)
```

**FILLING THE MARRIED MISSING VALUES BY THE MODE**

```python
dataset["Married"].fillna(dataset["Married"].mode()[0],inplace=True)
```

**for one hot encoding we have 2 methods first is get dummies with the help of get dummies we can perform the one hot encoding (it is method of pandas) and secound is scikit-learn library so in scikit-learn library we have one hot encoder class, with the help of this we can perform one hot encoding**

```python
en_data=dataset[["Gender","Married"]]
en_data
```

```python
pd.get_dummies(en_data)
```

**We can check the data type of en_data by info.**

```python
pd.get_dummies(en_data).info()
```

**from sklearn module there is preprocessing file and import onehotEncoder class**

```python
from sklearn.preprocessing import OneHotEncoder
```

**call the onehotencoder class through the ohe variable through this variable fit_transform it works as it see the data, understand the data and apply scikit-learn algorithm and it transform the data **

**Sparse_matrix: it have the element (0,1)**

```python
ohe=OneHotEncoder(drop="first")
ar=ohe.fit_transform(en_data).toarray()
```

**converting it into dataframe**

```python
pd.DataFrame(ar,columns=["Gender_Male "," Married_Yes "])
```
