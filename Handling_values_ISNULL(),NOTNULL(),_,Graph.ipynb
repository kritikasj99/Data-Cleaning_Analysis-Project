**IMPORTING LIBRARY**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

**LOADING DATASET**

```python
dataset = pd.read_csv("/content/loan.csv")
```

**For viewing head of the dataset**

```python
dataset.head()
```

**For total no.of Rows and Columns in the dataset**

```python
dataset.shape
```

**Null values**

```python
dataset.isnull()
```

**NOT NULL**

```python
dataset.notnull()
```

**Null values in each columns**

```python
dataset.isnull().sum()
```

**Not null in each columns**

```python
dataset.notnull().sum()
```

**To view % of null values in each columns**

```python
(dataset.isnull().sum()/(dataset.shape[0]))*100
```

**To view % of notnull values in each columns**

```python
(dataset.notnull().sum()/(dataset.shape[0]))*100
```

**Another method to view % of null values in each columns**

```python
(dataset.isnull().sum()/len(dataset))*100
```

**To view total null values in the dataset**

```python
dataset.isnull().sum().sum()
```

**To view total notnull vales in the dataset**

```python
dataset.notnull().sum().sum()
```

**To find the % of total null values in the dataset**

```python
(dataset.isnull().sum().sum()/(dataset.shape[0]*dataset.shape[1]))*100
```

**To find the % of total notnull values in the dataset**

```python
(dataset.notnull().sum().sum()/(dataset.shape[0]*dataset.shape[1]))*100
```

**viewing the graph of isnull values in the dataset**

```python
sns.heatmap(dataset.isnull())
plt.show()
```

**Viewing the graph of notnull values in the dataset**

```python
sns.heatmap(dataset.notnull())
plt.show()
```
