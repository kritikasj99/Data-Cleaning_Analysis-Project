**From sklearn library import the class which is simpleimputer**

```python
from sklearn.impute import SimpleImputer
```

**Through SI variable call the simpleimputer**

```python
si=SimpleImputer()
```

**Through SI class we call the strategy="mean" and through si class we get si.fit_transform()through this we can change the dataset into mean**

```python
si=SimpleImputer(strategy="mean")
si.fit_transform(dataset[['CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term',
       'Credit_History']])

```

**Converting the array into dataframe**

```python
si=SimpleImputer(strategy="mean")
ar=si.fit_transform(dataset[['CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term',
       'Credit_History']])
```

**Through pandas we can view the dataframe**

```python
pd.DataFrame(ar,columns=dataset.select_dtypes(include="float64").columns)
```

**Giving the variable name to the dataset**

```python
new_dataset=pd.DataFrame(ar,columns=dataset.select_dtypes(include="float64").columns)
```
