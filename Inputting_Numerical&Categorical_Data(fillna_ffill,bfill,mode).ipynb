**(FILLNA)-using ffill in categorical data**

****

```python
dataset["Gender"].fillna(dataset["Gender"].ffill(),inplace=True)
```

**(FILLNA)-using bfill in categorical data**

```python
dataset["Married"].fillna(dataset["Married"].bfill(),inplace=True)
```

**To view the object data type column in the dataset**

****

```python
dataset.select_dtypes(include="object").columns
```

**RUNNING FOR LOOP FOR OBJECT COLUMNS AND FILLING MISSING VALUES BY(FILLNA)-BFILL**

```python
for i in dataset.select_dtypes(include="object").columns:
  dataset[i].fillna(dataset[i].bfill(),inplace=True)
```

**RUNNING FOR LOOP FOR OBJECT COLUMNS AND FILLING MISSING VALUES BY (FILLNA)-FFILL**

```python
for i in dataset.select_dtypes(include="object").columns:
  dataset[i].fillna(dataset[i].ffill(),inplace=True)
```

**To calculate mode in gender**

```python
dataset["Gender"].mode()
```

**To view only male**

```python
dataset["Gender"].mode()[0]
```

**filling the mode in the gender column**

```python
dataset["Gender"].fillna(dataset["Gender"].mode()[0],inplace=True)
```

**To view the object data type column in the dataset**

```python
dataset.select_dtypes(include="object").columns
```

**Run for loop for the object data types by filling mode value in the missing values of object datatype **

```python
for i in dataset.select_dtypes(include="object").columns:
  dataset[i].fillna(dataset[i].mode()[0],inplace=True)
```
